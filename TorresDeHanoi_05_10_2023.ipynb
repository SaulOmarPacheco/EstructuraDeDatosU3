```python
class Pilas():


    def __init__(self):
            #Se crea una pila vacia
        self.A=[]
        self.B=[]
        self.C=[]
        self.lista=[]
    def Inicio(self):
        for i in range(1,5):
            self.A.append(i)

    def mostrar(self):
        for item in self.A:
            print(item)

    def juegoView(self):
        self.lista=[]
        self.lista.append(self.A)
        self.lista.append(self.B)
        self.lista.append(self.C)
        print(self.lista)
        print("\n---------------------------------")

    def mover(self):
        if len(self.C)==0 and len(self.B)==0:
            self.B.insert(0,self.A.pop(0))
            self.juegoView()
            self.mover()
    #Paso numero 1
        elif len(self.C)==0 and self.B[0]==1 and self.A[0]==2:
            self.C.insert(0,self.A.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 2
        elif len(self.A)==2 and len(self.B)==1 and len(self.C)==1 and self.B[0]==1 and self.C[0]==2:
            self.C.insert(0,self.B.pop(0))
            self.juegoView()
            self.mover()
    #Paso numero 3
        elif len(self.A)==2 and len(self.C)==2 and self.A[0]==3:
            self.B.insert(0,self.A.pop(0))
            self.juegoView()
            self.mover()
    #Paso numero 4
        elif len(self.A)==1 and len(self.B)==1 and len(self.C)==2 and self.A[-1]==4 and self.B[-1]==3 and self.C[0]==1:
            self.A.insert(0,self.C.pop(0))
            self.juegoView()
            self.mover()
    #Paso numero 5
        elif len(self.A)==2 and len(self.B)==1 and len(self.C)==1 and self.A[0]==1 and self.B[-1]==3 and self.C[-1]==2:
            self.B.insert(0,self.C.pop(0))
            self.juegoView()
            self.mover()
    #Paso numero 6
        elif len(self.A)==2 and len(self.B)==2 and len(self.C)==0 and self.A[0]==1 and self.B[0]==2:
            self.B.insert(0,self.A.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 7
        elif len(self.A)==1 and len(self.B)==3 and len(self.C)==0 and self.A[-1]==4 and self.B[0]==1:
            self.C.insert(0,self.A.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 8
        elif len(self.A)==0 and len(self.B)==3 and len(self.C)==1 and self.B[0]==1 and self.C[-1]==4:
            self.C.insert(0,self.B.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 9
        elif len(self.A)==0 and len(self.B)==2 and len(self.C)==2 and self.B[0]==2 and self.C[0]==1:
            self.A.insert(0,self.B.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 10
        elif len(self.A)==1 and len(self.B)==1 and len(self.C)==2 and self.A[-1]==2 and self.B[-1]==3 and self.C[0]==1:
            self.A.insert(0,self.C.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 11
        elif len(self.A)==2 and len(self.B)==1 and len(self.C)==1 and self.A[0]==1 and self.B[-1]==3 and self.C[-1]==4:
            self.C.insert(0,self.B.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 12
        elif len(self.A)==2 and len(self.B)==0 and len(self.C)==2 and self.A[0]==1 and self.C[0]==3:
            self.B.insert(0,self.A.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 13
        elif len(self.A)==1 and len(self.B)==1 and len(self.C)==2 and self.A[-1]==2 and self.B[-1]==1 and self.C[0]==3:
            self.C.insert(0,self.A.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 14
        elif len(self.A)==0 and len(self.B)==1 and len(self.C)==3 and self.B[-1]==1 and self.C[0]==2:
            self.C.insert(0,self.B.pop(0))
            self.juegoView()
            self.mover()
        #Paso numero 15
        else:
            print("Aacaba")


p=Pilas()
p.Inicio()
p.juegoView()
p.mostrar()
p.mover()
```
